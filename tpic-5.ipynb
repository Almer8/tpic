```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder

from sklearn.tree import DecisionTreeClassifier, plot_tree

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import classification_report, confusion_matrix
```

```python
dataset = pd.read_csv("/kaggle/input/crop-recommendation-dataset/Crop_recommendation.csv")
```

```python
X = dataset.drop("label", axis = 1)
y = dataset["label"]

le = LabelEncoder()
y_encoded = le.fit_transform(y)

class_names = le.classes_
print(f"{len(class_names)} класів")
print(class_names)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.3, random_state=42, stratify=y_encoded)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
dt_model_1 = DecisionTreeClassifier(max_depth=3, random_state=42)
dt_model_1.fit(X_train_scaled, y_train)
```

```python
dt_model_2 = DecisionTreeClassifier(max_depth=7, random_state=42)
dt_model_2.fit(X_train_scaled, y_train)
```

```python
dt_model_3 = DecisionTreeClassifier(max_depth=None, random_state=42)
dt_model_3.fit(X_train_scaled, y_train)
```

```python
y_pred_1 = dt_model_1.predict(X_test_scaled)
y_pred_2 = dt_model_2.predict(X_test_scaled)
y_pred_3 = dt_model_3.predict(X_test_scaled)
```

```python
print(classification_report(y_test, y_pred_1, target_names=class_names))
```

```python
print(classification_report(y_test, y_pred_2, target_names=class_names))
```

```python
print(classification_report(y_test, y_pred_3, target_names=class_names))
```

```python
models = {
    "Depth=3": (dt_model_1, y_pred_1),
    "Depth=7": (dt_model_2, y_pred_2),
    "Depth=None": (dt_model_3, y_pred_3)
}

results = []
for name, (model, y_pred) in models.items():
    train_acc = model.score(X_train_scaled, y_train)
    test_acc = accuracy_score(y_test, y_pred)
    results.append({
        "Model": name,
        "Train Accuracy": train_acc,
        "Test Accuracy": test_acc,
        "Difference": train_acc - test_acc
    })

results_df = pd.DataFrame(results)
display(results_df)
```

```python
plt.figure(figsize=(25, 15)) 
plot_tree(dt_model_1, 
          feature_names=X.columns,
          class_names=class_names,
          filled=True, 
          rounded=True,
          fontsize=10)
plt.show()
```

```python
cm = confusion_matrix(y_test, y_pred_2)

plt.figure(figsize=(18, 15))
sns.heatmap(cm, 
            annot=True, 
            fmt='d', 
            cmap='Blues',
            xticklabels=class_names, 
            yticklabels=class_names)

plt.xlabel('Predicted Label', fontsize=14)
plt.ylabel('True Label', fontsize=14)
plt.show()
```
