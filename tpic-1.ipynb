```python
from experta import *
```

```python
class Crop(Fact):
    nitrogen = Field(float, default=0.0)
    phosphorous = Field(float, default=0.0)
    potassium = Field(float, default=0.0)
    temperature = Field(float, default=0.0)
    humidity = Field(float, default=0.0)
    ph = Field(float, default=0.0)
    rainfall = Field(float, default=0.0)
    pass
```

```python
class CropEngine(KnowledgeEngine):
    @Rule(Crop(rainfall=P(lambda r: r<=30)))
    def muskmelon(self):
        print("Plant muskmelon")

    @Rule(Crop(rainfall=P(lambda r: r > 30),
               humidity=P(lambda h: h <= 27.7),
               potassium=P(lambda p: p <= 50)))
    def kidneybeans(self):
        print("Plant kidneybeans")

    @Rule(Crop(rainfall=P(lambda r: r > 30),
               humidity=P(lambda h: h <= 27.7),
               potassium=P(lambda p: p > 50)))
    def chickpea(self):
        print("Plant chickpea")

    @Rule(Crop(rainfall=P(lambda r: r <= 82.1),
               humidity=P(lambda h: 27.7 < h <= 70.8),
               phosphorous=P(lambda p: p <= 107.5),
               nitrogen=P(lambda n: n <= 59.5)))
    def lentil(self):
        print("Plant lentil")

    @Rule(Crop(rainfall=P(lambda r: r > 82.1),
               humidity=P(lambda h: 27.7 < h <= 70.8),
               phosphorous=P(lambda p: p <= 107.5),
               nitrogen=P(lambda n: n <= 59.5)))
    def mango(self):
        print("Plant mango")

    @Rule(Crop(rainfall=P(lambda r: r <= 112.5),
               humidity=P(lambda h: 27.7 < h <= 70.8),
               phosphorous=P(lambda p: p <= 107.5),
               nitrogen=P(lambda n: n > 59.5)))
    def maize(self):
        print("Plant maize")

    @Rule(Crop(rainfall=P(lambda r: r > 112.5),
               humidity=P(lambda h: 27.7 < h <= 70.8),
               phosphorous=P(lambda p: p <= 107.5),
               nitrogen=P(lambda n: n > 59.5)))
    def coffee(self):
        print("Plant coffee")

    @Rule(Crop(humidity=P(lambda h: h > 70.8),
               phosphorous=P(lambda p: p <= 32.5),
               potassium=P(lambda p: p <= 20)))
    def orange(self):
        print("Plant orange")

    @Rule(Crop(humidity=P(lambda h: h > 70.8),
               phosphorous=P(lambda p: p <= 32.5),
               potassium=P(lambda p: p > 20)))
    def coconut(self):
        print("Plant coconut")

    @Rule(Crop(humidity=P(lambda h: 27.7 < h <= 90),
               phosphorous=P(lambda p: 32.5 < p <= 107.5)))
    def banana(self):
        print("Plant banana")

    @Rule(Crop(humidity=P(lambda h: h > 90),
               phosphorous=P(lambda p: 32.5 < p <= 107.5)))
    def papaya(self):
        print("Plant papaya")

    @Rule(Crop(humidity=P(lambda h: h > 27.7),
               phosphorous=P(lambda p: p > 107.5),
               rainfall=P(lambda r: r <= 87.5)))
    def grapes(self):
        print("Plant grapes")

    @Rule(Crop(humidity=P(lambda h: h > 27.7),
               phosphorous=P(lambda p: p > 107.5),
               rainfall=P(lambda r: r > 87.5)))
    def apple(self):
        print("Plant apple")
```

```python
engine = CropEngine()
engine.reset()
engine.declare(Crop(nitrogen=14.0, 
                    phosphorous=19.0, 
                    potassium=14.0,
                    temperature=17.68408797,
                    humidity=94.35815354,
                    ph=6.699164936,
                    rainfall=108.0638166))
engine.run()
```
